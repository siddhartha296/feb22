```python
#subset column with iloc
#iloc allows us to use integers
yr_p_subset_iloc=df1.iloc[:,[2,4,-1]]
print(yr_p_subset_iloc)# we are subsetting 2,4, and last columns that is [2,4,-1]
```

```python
#subsetting using range
small_range=list(range(4))
small_range_iloc=df1.iloc[:,small_range]
print(small_range_iloc)
```

```python
l=[3,5]#place the index of columns you want to parse in this  list and use iloc with ":"
df1.iloc[:,l]
```

```python
#subsetting both rows and columns
#using loc
df1.loc[42,'country']# what is the countryin 42 row?
```

```python
#subsetting multiple rows and columns at the same time
df1.iloc[[0,9,99],[0,3,4]]
```

Grouped means

```python
#for each year in our data what is the avg life expectancy ?
df1.groupby('year')['lifeExp'].mean()
```

```python
#multiple group by
multi_group_var=df1.groupby(['year','continent'])['lifeExp','gdpPercap'].mean()
print(multi_group_var)
```

```python
#flatening the table to the initial form if want this view
#we use reset-index() command to do this
flatened_table=multi_group_var.reset_index()
print(flatened_table)
```

grouped frequency counts


```python
#we can count the unique values using nunique command
#here we are finding the no of unique countries in each continent
df1.groupby('continent')['country'].nunique()
```

**Some basic plots**

```python
global_yearly_lifeExp=df1.groupby('year')['lifeExp'].mean()
print(global_yearly_lifeExp)
```

```python
#when the year increases the life expectancy also increases
global_yearly_lifeExp.plot()
```
